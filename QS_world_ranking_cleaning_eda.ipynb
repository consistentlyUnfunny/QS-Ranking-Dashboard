# Creating new columns and dropping unecessary columns

```python
def extract_num_ranking(value):
  if pd.isna(value):
    return None
  if '+' in value:
    return int(value.replace('+', '')) # replace tailing plus symbol to blank
  if '-' in value:
    return int(value.split('-')[0]) # get the upper bound of the range
  if '=' in value:
    return int(value.replace('=', ''))
  return int(value)
```

```python
# Create new columns for numeric values of the rankings
rankings['RANK_2025_NUM'] = rankings['RANK_2025'].apply(extract_num_ranking)
rankings['RANK_2024_NUM'] = rankings['RANK_2024'].apply(extract_num_ranking)

# Calculate rank change
rankings['RANK_CHANGED'] = rankings['RANK_2025_NUM'] - rankings['RANK_2024_NUM']
```

```python
# Convert other ranking columns
rank_cols = [
    'Academic_Reputation_Rank',
    'Employer_Reputation_Rank',
    'Faculty_Student_Rank',
    'Citations_per_Faculty_Rank',
    'International_Faculty_Rank',
    'International_Students_Rank',
    'International_Research_Network_Rank',
    'Employment_Outcomes_Rank',
    'Sustainability_Rank'
]

for col in rank_cols:
  rankings[col + '_Num'] = rankings[col].apply(extract_num_ranking)
```

```python
rankings.info()
```

```python
rankings.to_csv("QS Ranking Cleaned.csv", index = False)
```

# Simple Visualization using Matplotlib and Seaborn

```python
import pandas as pd
rankings = pd.read_csv('/content/QS Ranking Cleaned.csv')
rankings.head()
```

## Framework for matplotlib + seaborn
1. Import libraries  
2. Load and prep data  
3. Create figure + axis (optional with Seaborn)  
4. Call sns.plot() or ax.plot()  
5. Customize and show plot

## Top 20 universities by Overall Score

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Organize data
top_overall = rankings.dropna(subset=['Overall_Score']).sort_values(by='Overall_Score', ascending = False).head(20)

# Define fig and ax
fig, ax = plt.subplots(figsize=(8, 10))

# Plot and Customize
colors = sns.color_palette('magma', n_colors = 20)
ax.barh(top_overall['Institution_Name'], top_overall['Overall_Score'], color = colors)
ax.set_xlabel("Overall Score")
ax.set_ylabel("Institution")
ax.set_title("Top 20 Universities by Overall Score")

plt.gca().invert_yaxis() # get current axis (gca) and invert the order
plt.show()
```

## Best and Worst Ranking Change

```python
sorted_rank_change = rankings.sort_values(by='RANK_CHANGED', ascending=False)
best_rank_change = sorted_rank_change.head(10)
worst_rank_change = sorted_rank_change.dropna(subset='RANK_CHANGED').tail(10)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize = (16, 10))
colors1 = sns.color_palette("mako", n_colors = 10)
colors2 = sns.color_palette("rocket", n_colors = 10)

ax[0].barh(best_rank_change['Institution_Name'], best_rank_change['RANK_CHANGED'], color = colors1)
ax[0].set_xlabel("Rank Changed")
ax[0].set_ylabel("Institution")
ax[0].set_title("Best Ranking Change")
ax[0].invert_yaxis()

ax[1].barh(worst_rank_change['Institution_Name'], worst_rank_change['RANK_CHANGED'], color = colors2)
ax[1].set_xlabel("Rank Changed")
ax[1].set_ylabel("Institution")
ax[1].set_title("Worst Ranking Change")
ax[1].invert_yaxis()

fig.suptitle("Best and Worst Ranking Change", fontsize= 16)

plt.tight_layout()
plt.show()
```

## University Distribution According to Region

```python
rankings['Region'].unique()
```

```python
# Count number of universities in each region
region_counts = rankings['Region'].value_counts()

# Color palette
colors = sns.color_palette("husl", n_colors = len(region_counts))

# Create donut chart
fig, ax = plt.subplots(figsize = (8, 8))
wedges, texts, autotexts = ax.pie(
    region_counts.values,
    labels = None,
    autopct = '%1.1f%%',
    colors = colors,
    startangle = 90,
    wedgeprops = dict(width = 0.4), # Creates donut hole
    pctdistance = 1.1 # push the labels outward
)


for t in autotexts:
    t.set_fontsize(10)

ax.set_title("University Distribution by Region")
ax.legend(wedges, region_counts.index, title = "Region", bbox_to_anchor=(1, 0.5))
plt.show()
```

## Correlation matrix

```python
# Select only numeric columns
score_cols = [col for col in rankings.columns if 'Score' in col ]
corr_data = rankings[score_cols]

# .corr() to get Pearson correlation
corr_matrix = corr_data.corr()

# Visualize
plt.figure(figsize = (12, 10))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt = '.2f', square = True, linewidths = 0.5)
plt.title("Correlation Matrix of Score Metrics")
plt.show()
```
